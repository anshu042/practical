###Exploratory Data Analysis (EDA)

```python
#relationship between Symboling and price
plt.figure(figsize=(20,8))
plt.title('Symboling vs Price')
sns.boxplot(x=df.symboling, y=df.price)
plt.show()
```

```python
#fueltype vs price
sns.barplot(data=df, x='fueltype', y='price')
plt.title('fueltype vs price')
```

```python
#relation between carbody vs horsepower
sns.lineplot(data=df, x='carbody', y='horsepower')
plt.title('carbody vs horsepower')
```

```python
plt.figure(figsize=(5,100))
df['aspiration'].value_counts().plot.pie(labeldistance=None)
plt.legend(loc='best', bbox_to_anchor=(-0.1,1.))
plt.title('cars aspiration')
```

```python
plt.figure(figsize=(15,5))
sns.histplot(data=df, x='carbody',bins=10, color='red')
```

Univariate analysis is performed in above plot, which displays the total
count of carbody.

```python
#creation of the heatmap
plt.figure(figsize=(20,8))
cor=df.corr()
sns.heatmap(cor, xticklabels=cor.columns, yticklabels=cor.columns, annot=True)
```
