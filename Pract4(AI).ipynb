```python
df['Date'] = pd.to_datetime(df['date_issued'])
df['Month'] = df['Date'].dt.month
df['date_issued:year'] = df['Date'].dt.year
```

```python
df['Month']
```

```python
df['date_issued:month'] = df['Date'].dt.month
df['date_issued:day'] = df['Date'].dt.day
df['date_issued:day_of_week'] = df['Date'].dt.day_of_week
df['date_issued:day_of_year'] = df['Date'].dt.day_of_year
```

```python
df[['date_issued','date_issued:year','date_issued:month','date_issued:day','date_issued:day_of_week','date_issued:day_of_year']].head()
```

```python
def week_part(day):
  if day in [1,2,3,4,5,6,7]:
    return "Week 1"
  elif day in [8,9,10,11,12,13,14]:
    return "Week 2"
  elif day in [15,16,17,18,19,20,21]:
    return "Week 3"
  elif day in [22,23,24,25,26,27,28]:
    return "Week 4"
  elif day in [29,30,31]:
    return "Week 5"
```

```python
# utilize it along with apply method
df['date_issued:hour'] = df['Date'].dt.hour
df['date_issued:Week_No'] = df['date_issued:day'].apply(week_part)
df.head(20)
```

```python
import numpy as np
df['date_issued:is_weekend'] = np.where(df['date_issued:day_of_week'].isin([5,6]),1,0)
df[['date_issued','date_issued:day_of_week','date_issued:is_weekend']].head()
```

```python
df.drop('date_issued:hour',axis=1)
```

```python
df.info()
```

```python
import numpy as np
df['log_date_issued:year']=np.log(df['date_issued:year'])

df['log_date_issued:year']
```

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('supermarket_sales.csv')
df.head()
```

```python
df.head()
```

###Numeric Aggregations

```python
grouped_df = df.groupby('Branch')
df[['tax_branch_mean','unit_price_mean']] = grouped_df[['Tax 5%','Unit price']].transform('mean')
df[['tax_branch_std','unit_price_std']] = grouped_df[['Tax 5%','Unit price']].transform('std')
df[['product_count','gender_count']] = grouped_df[['Product line','Gender']].transform('count')
```

```python
df[['Branch','tax_branch_mean','unit_price_mean','tax_branch_std','unit_price_std','product_count','gender_count']].head(10)
```

```python
df[['Tax 5%','Unit price','Branch','tax_branch_mean','unit_price_mean']]
```

###Indicator Variables and Interaction Terms

```python
import numpy as np
df['unit_price_50'] = np.where(df['Unit price'] > 50,1,0)
df['unit_price_50*qty'] = df['unit_price_50']*df['Quantity']
```

```python
df[['unit_price_50','unit_price_50*qty']].head()
```

###Numeric Transformations

```python
df['log_cogs']=np.log(df['cogs']+1)
df['gross income squared']=np.square(df['gross income'])
```

```python
df[['cogs','log_cogs','gross income','gross income squared']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,6))
sns.histplot(df['cogs'], ax=ax1, kde=True)
sns.histplot(df['log_cogs'], ax=ax2, kde=True);
```
